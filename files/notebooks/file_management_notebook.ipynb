```python
expense_company_set = set()
expense_type_set = set()
for year in sorted(expense_report_dict.keys()):
    for month in sorted(expense_report_dict[year].keys()):
        for expense_report in expense_report_dict[year][month]:
            expense_date = expense_report.get_date()
            expense_type = expense_report.get_type()
            expense_company = expense_report.get_company()
            if(expense_type not in expense_type_set):
                #print("First occurance of TYPE=", expense_type, " at DATE=", expense_date)
                expense_type_set.add(expense_type)
            if(expense_company not in expense_company_set):
                #print("First occurance of COMPANY=", expense_company, " at DATE=", expense_date)
                expense_company_set.add(expense_company)
print("\nexpense_type_set:", expense_type_set)
#print("\nexpense_company_set:", expense_company_set)
```
